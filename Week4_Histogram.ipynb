```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("deneme.jfif")
```

```python
cv.imshow('image',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = http://localhost:8888/notebooks/Week4_Histogram.ipynb#'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```
